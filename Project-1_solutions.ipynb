# Problem Statement

#### 4. What is the hypothesis?

**Answer:** GRE score, GPA, and rank of a particular student's alma mater can predict whether or not the student will be admitted into graduate school at UCLA with a success rate of 70% (assumed percentage).

#### 5. What are the assumptions for this problem?

**Assumptions** 
1. The variables in our data set are valid determiners as to whether or not a student will be accepted into graduate school.

2. A 70% success rate is sufficient for the purposes of our analysis

3. Other variables, such as time/date of application, age, gender, and department are not important

4. The data was collected in an un-biased manner

5. We are ignoring factors that have nothing to do with the student, such as the budget of the graduate school

#### 6. Whare are the risks of our data?

1. The variables we have identified may not be sufficient to predict admission status of a student.
2. We may have too many variables - thus over-constraining our model.
3. Some students may have been admitted due to external factors, such as through connections within their network, thus applying "dead weight" to our model and poorly constraining it.
4. Because the data is cross-sectional, we are unable to determine if time has a significant influence on whether or not a student is admitted (i.e. if the student was rejected due to the department experiencing budget cuts during a recession).
5. Building on Assumption 4, we know nothing about when the data was collected. If this data was collected in 1980, it may no longer be relevent to our analysis.

# Analysis

#### Eploratory Analysis

To explore our data, we can employ a number of different methods to get a feel for what variables are important to our predictions. 

For a quick visualization, we can render the data in a scatter plot and get a sense of where the admitted students tend to lie.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline  
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

for n in range(0,len(df)):
    xs = df.gpa[n]
    ys = df.gre[n]
    zs = df.prestige[n]
    
    if df.admit[n] == 1:
        c = 'r'
    else:
        c = 'b'
        
    ax.scatter(xs,ys,zs,c=c,marker='o')
    
ax.set_xlabel('gpa')
ax.set_ylabel('gre')
ax.set_zlabel('prestige')

plt.show()
```
